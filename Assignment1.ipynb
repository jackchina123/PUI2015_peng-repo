```python
from __future__ import print_function
__author__= 'fbb'
import os
from scipy.optimize import curve_fit, minimize
import numpy as np
import pylab as pl
import json

distributions = ['pois', 'gaus', 'chisq', 'cauchy', 'lnorm', 'binomial']
df = 100
md={}


mysize=1000/(np.array(range(1,100)+[10]))
#print (mysize, mysize.shape)

#cauchy mean


import json
s = json.load( open('fbb_matplotlibrc.json') )
pl.rcParams.update(s)

%pylab inline
```

```python
md['chisq']={} 
for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size=n)

    
means = {}
means['chisq'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('chisq sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

allmeans=[]
for n in means['chisq']:
    allmeans.append(means['chisq'][n])
pl.figure(figsize=(10,10))
pl.hist(allmeans, bins=30)
pl.xlabel('chisq sample mean')
pl.ylabel('N')
```

```python
md['gaus']={} 
for n in mysize:
    md['gaus'][n] = np.random.normal(100, 7, 100)

    
means = {}
means['gaus'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['gaus'].iterkeys():
    means['gaus'][nn] = md['gaus'][nn].mean()
    axchisq_mu_n.plot(nn, means['gaus'][nn], 'o')
    axchisq_mu_n.set_xlabel('gaus sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

allmeans=[]
for n in means['gaus']:
    allmeans.append(means['gaus'][n])
pl.figure(figsize=(10,10))
pl.hist(allmeans, bins=30)
pl.xlabel('gaus sample mean')
pl.ylabel('N')
```
